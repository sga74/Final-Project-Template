## Introduction
Like many other coastal locations across the globe, Califonia's coastal cities are experienceing the effects of sea level rise caused by climate change. In many instances this manifests as an increase in flooding events and accelerated coastal erosion. However, one factor that makes California unique is its geology. There are varying rates of vertical land motion up and down Califonia's coast that can either reduce or intensify the effects of sea level rise depending on location. In some locations there is vertical uplift that can reduce the rate of sea level rise, and in others there is sinking that can increase it. In this project, I examined the rates of sea level rise at three different coastal locations in California that represent the North, South, and center of the coast across a ~25yr period (from 1992-2017). I predicted that my Northern location (Humboldt Bay) would show the greatest rate of sea level rise due to the sinking plate beneath it and becuase it is projected to see the most rapid sea level rise in the state in the coming decades.

## 1. Producing a yearly mean sea level graph for each individual location  

I loaded the monthly SSH data from the NASA ECCO v4r4 dataset that I had previously narrrowed to each study location by selecting coordinates in the sea level portal. I also resampled data to aquire the yearly mean and converted it out of datetime in order to plot each site and include a regression line. 

```python
# load data file(s)
# I started with the La Jolla file, most of my troublshooting was done here so I could repeat with other locations
s_df = pd.read_csv('LAJollassh.csv',
                  parse_dates=['DateTime'],index_col=['DateTime']) 
# Read in the file and made sure the times were being read as datetime objects and were set to be the index

SLJ=s_df['Sea Surface Height (ECCO v4r4)'] # assigned variable for unedited sea surface heights
LJ_Annual= SLJ.resample("YS").mean() # resampled the unedited sea surface heights to get the yearly mean 
annlj_df = LJ_Annual.to_frame() # made a new data fram with yearly sea surface heigths + assigned new variable

#linregress apparently does not like datetime so now coverting the time to integers with .astype
# Received help from Sage, Lydia, and stackxchange woth troubleshooting this step
s_df['time']=s_df.index.astype(np.int64)//10**9
annlj_df['time']=annlj_df.index.astype(np.int64)//10**9
# also assigned new variable representing the yealry La Jolla converted times and a new column (time) for plotting

x_mlj=s_df['time'] 
y_mlj=SLJ
x_ylj=annlj_df['time']
y_ylj=LJ_Annual
# created plotting variables that represent monthly time, yearly time, monthly sea level, and yearly sea level.
# At this point I did not know what would be better suited for my data so I plotted both monthly and yearly

lxslope,lxintercept, lxr_value, lxp_value, lxstd_err = stats.linregress(x_mlj, y_mlj)
lyslope,lyintercept, lyr_value, lyp_value, lystd_err = stats.linregress(x_ylj, y_ylj)
# performed linregress function on both monthly and yearly data

s_df['regression line'] = lxslope * s_df['time'] + lxintercept
annlj_df['regression line'] = lyslope * annlj_df['time'] + lyintercept
#used values obtained by linregress to produce regression line for monthly and yearly 

#adjusted the x axis for plotting now that regression line is complete 
s_df['DateTime']=pd.to_datetime(s_df.index)
annlj_df['DateTime']=pd.to_datetime(annlj_df.index)

#created a plot for monthly mean data 
fig, axs = plt.subplots()
axs.plot(s_df['DateTime'], y_mlj, marker= '.') #plotted data 
axs.plot(s_df['DateTime'], s_df['regression line'], marker= '.')#plotted regression line
axs.set_xlabel("Time (Months)", fontsize = 10)
axs.set_ylabel("SSH at La Jolla ", fontsize = 10)

#created plot for yearly data 
fig, axs = plt.subplots()
axs.plot(annlj_df['DateTime'], y_ylj, marker= '.') # data
axs.plot(annlj_df['DateTime'], annlj_df['regression line'], marker= '.') # regression line
axs.set_xlabel("Time (Years)", fontsize = 10,)
axs.set_ylabel("SSH at La Jolla ", fontsize = 10)
print(lxslope,lxr_value) 
print(lyslope, lyr_value)
# printed the slope and R^2 values for the regression lines so I could eventually compare between locations
```

```python
#These plots were just my first exploratory plots so see initial trends and compare monthly and yearly
# At this point I decided that I would only plot the yearly means for all subsequent locations
# Now repeat initial cleaning steps for Humboldt and San Francisco (no changes were necessary)
```

```python
#repeated cleaning steps with new variables to represent Humboldt data +excluding monthly steps
HB_df = pd.read_csv('SSHHumboldt.csv',
                  parse_dates=['DateTime'],index_col=['DateTime'])

SHB=HB_df['Sea Surface Height (ECCO v4r4)']
HB_Annual= SHB.resample("YS").mean()
annhb_df = HB_Annual.to_frame()

HB_df['time']=HB_df.index.astype(np.int64)//10**9
annhb_df['time']=annhb_df.index.astype(np.int64)//10**9

x_yhb=annhb_df['time']
y_yhb=HB_Annual

hyslope,hyintercept, hyr_value, hyp_value, hystd_err = stats.linregress(x_yhb, y_yhb)

annhb_df['regression line'] = hyslope * annhb_df['time'] + hyintercept

annhb_df['DateTime']=pd.to_datetime(annhb_df.index)

print(hyslope, hyr_value)
```

```python
#repeated cleaning steps with new variables to represent San Francisco data +excluding monthly steps
SF_df = pd.read_csv('SSHSF.csv',
                  parse_dates=['DateTime'],index_col=['DateTime'])

SSF=SF_df['Sea Surface Height (ECCO v4r4)']
SF_Annual= SSF.resample("YS").mean()
annsf_df = SF_Annual.to_frame()

SF_df['time']=SF_df.index.astype(np.int64)//10**9
annsf_df['time']=annsf_df.index.astype(np.int64)//10**9

x_ysf=annsf_df['time']
y_ysf=SF_Annual

syslope,syintercept, syr_value, syp_value, systd_err = stats.linregress(x_ysf, y_ysf)

annsf_df['regression line'] = syslope * annsf_df['time'] + syintercept

annsf_df['DateTime']=pd.to_datetime(annsf_df.index)
print(syslope, syr_value)
```

```python
#Now produce a plot with each location as a stacked subplot to serve as initial comparison of trends over time
fig, axs = plt.subplots(nrows=3, ncols=1, figsize = (10,15))
#three subplots stacked vertically 
axs[0].plot(annhb_df['DateTime'], y_yhb, marker= '.', linewidth = 2.2, color = 'darkorange', markersize = 12)
#plotted data assigning linewidth and color 
axs[0].plot(annhb_df['DateTime'], annhb_df['regression line'], color = 'red')
#plotted regression line 
axs[0].set_xlabel("Time (Years)", fontsize = 14,)
axs[0].set_ylabel("SSH at Humboldt Bay ", fontsize = 14)
axs[0].grid(True)

axs[1].plot(annlj_df['DateTime'], y_ylj, marker= '.', linewidth = 2.2, color = 'mediumseagreen', markersize=12)
axs[1].plot(annlj_df['DateTime'], annlj_df['regression line'], color = 'red')
axs[1].set_xlabel("Time (Years)", fontsize = 14)
axs[1].set_ylabel("SSH at La Jolla ", fontsize = 14)
axs[1].grid(True)


axs[2].plot(annsf_df['DateTime'], y_ysf, marker= '.', linewidth = 2.2, color = "mediumblue", markersize=12)
axs[2].plot(annsf_df['DateTime'], annsf_df['regression line'], marker= '.', color = "red")
axs[2].set_xlabel("Time (Years)", fontsize = 14,)
axs[2].set_ylabel("SSH at San Francisco ", fontsize = 14)
axs[2].grid(True)
```
